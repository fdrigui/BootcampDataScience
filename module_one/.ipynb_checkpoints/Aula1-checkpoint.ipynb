### 1.1 Importação do *valor total dos procedimentos hospitalares do SUS por UF e mês de **processamento***

**Informações da extração:**
- [**Informações de Saúde(TABNET)**](http://www2.datasus.gov.br/DATASUS/index.php?area=0202&id=11633&VObj=http://tabnet.datasus.gov.br/cgi/deftohtm.exe?sih/cnv/qi)
- **Localização:** Informações de Saúde(TABNET)/Assistência à Saúde/Dados Consolidados AIH (RD), por local de internação, a partir de 2008/Brasil por região e unidade de Federação
- **Data:** 18-05-2021 22:35
- **Linha:** Unidade da Federação
- **Coluna:** Ano Mês processamento
- **Conteúdo:** Valor Total
- **Período:** *de:* Jan-2008 *até:* Mar-2021
- **Nome do arquivo:** /dados/A223850189_28_143_208(proc).csv

````python
# Abre e Lê o arquivo csv
df_proc = pd.read_csv('https://raw.githubusercontent.com/fdrigui/BootcampDataScience/main/module_one/dados/A223850189_28_143_208(proc).csv',
                      sep=';', decimal=',', thousands='.', engine='python', encoding='ISO-8859-1', skiprows=3, skipfooter=12)

df_proc.head()
````

## 2. Preparação dos dados

### 2.1 Altera '/' por '-' no nome da coluna.
'/' não é um caracter muito comum para se ter nos nomes das colunas

````python
df_proc.columns = df_proc.columns.str.replace('/', '-')
````

### 2.2 Nome da coluna 'UNidade da Federação'
A Coluna com o nome 'Unidade da Federação' é demazindo grande e com acento.<br>
Renomear de: ```Unidade da Federação``` para: ```uf_name```

````python
df_proc = df_proc.rename(columns={'Unidade da Federação':'uf_name'})
````

### 2.2 Divide ```uf_name``` em ```código``` e ```unidade federativa```

````python
df_proc[['uf_code', 'UnidadeFederativa']] = df_proc['uf_name'].str.split(' ',n=1, expand=True).rename(columns={0:'uf_code', 1:'UnidadeFederativa'})
````

### 2.3 Cria Dataframe de código de UF e UF
para isso nós vamos utilizar uma lista de referencia utilizada [LINK](https://www.oobj.com.br/bc/article/quais-os-c%C3%B3digos-de-cada-uf-no-brasil-465.html)

````python
# Cria lista na mão dos códigos UF e os UFs
uf_number_code = pd.DataFrame({"uf_number":["11", "12", "13", "14", "15", "16", "17", "21", "22", "23", "24", "25", "26", "27",
                                          "28", "29", "31", "32", "33", "35", "41", "42", "43", "50", "51", "52", "53"],
                               "uf":["RO", "AC", "AM", "RR", "PA", "AP", "TO", "MA", "PI", "CE", "RN", "PB", "PE", "AL",
                                       "SE", "BA", "MG", "ES", "RJ", "SP", "PR", "SC", "RS", "MS", "MT", "GO", "DF"]})
````

### 2.4 Cria coluna 'UF' com a abreviação 

````python
df_proc['uf'] = df_proc['uf_code'].replace(uf_number_code['uf_number'].array,uf_number_code['uf'].array)
df_proc['uf_i'] = df_proc['uf']
````

### 2.5 Define ```df_proc['uf']``` como indice

````python
df_proc = df_proc.set_index('uf_i')
df_proc.head()
````

````python
def plot_bar(xaxis, yaxis, data, title, xlabel, ylabel):
    plt.figure(figsize=(15,10))
    ax = sns.barplot(x=xaxis, y=yaxis, data=data.sort_values(by=yaxis, ascending=False), palette='rocket')
    ax.set_title(title, fontsize=18, loc='left')
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ticks_y = ticker.FuncFormatter(lambda x, pos: '{0:g}'.format(x/1e6))
    ax.yaxis.set_major_formatter(ticks_y)
    plt.show()
````

````python
plot_bar(xaxis='uf', yaxis='2008-Jan', data=df_proc,
         title='Valor total de procedimentos processados do SUS no mes de Janeiro de 2008 por estado',
         xlabel='Unidades Federativas', ylabel='Valor total processado (R$ Milhões)')
````

````python
plot_bar(xaxis='uf', yaxis='2021-Mar', data=df_proc,
         title='Valor total de procedimentos processados do SUS no mes de Janeiro de 2008 por estado',
         xlabel='Unidades Federativas', ylabel='Valor total processado (R$ Milhões)')
````
